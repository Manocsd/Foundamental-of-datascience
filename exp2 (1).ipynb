```python
df=pd.read_csv(r"C:\Users\HDC0422087\Downloads\inventory_dataset.csv")
```

```python
null_columns=df.columns[df.isnull().any()]
```

```python
null_columns
```

```python
median_purchase_price = df['Purchase_Price'].median()
df['Purchase_Price'].fillna(median_purchase_price)
df
```

```python
mean=df['Sale_Price'].mean()
```

```python
df['Sale_Price'].fillna(mean)
df
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
Q1 = df['Sale_Price'].quantile(0.25)
Q3 = df['Sale_Price'].quantile(0.75)
IQR = Q3 - Q1
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
outliers = df[(df['Sale_Price'] < lower_bound) | (df['Sale_Price'] > upper_bound)]
print("Number of outliers:", len(outliers))
```

```python
df = df[(df['Sale_Price'] >= lower_bound) & (df['Sale_Price'] <= upper_bound)]
```

```python
df
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
df.loc[:,'Quantity_In_Stock'] = scaler.fit_transform(df[['Quantity_In_Stock']])
df
```

```python
df.loc[:, 'Quantity_In_Stock'] = (df['Quantity_In_Stock'] * 100).astype(int)
```

```python
df
```

```python
df.loc[:,'Discontinued'] = df['Discontinued'].astype(int)
```

```python
df
```

```python
df.to_csv('cleaned_inventory.csv', index=False)
```
